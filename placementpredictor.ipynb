```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/collegePlace.csv")
df.head(5)
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Stream']=le.fit_transform(df['Stream'])
df.dtypes
```

```python
for i in df.columns:
  df[i]=df[i].astype(int)
df.dtypes
```

```python
df.corr()
```

```python
x=df[['Internships','CGPA','Hostel','HistoryOfBacklogs']]
y=df['PlacedOrNot']
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="gini",random_state=1,max_depth=3,min_samples_leaf=2)
```

```python
from sklearn.metrics import accuracy_score
model.fit(x_train,y_train)
predict2=model.predict(x_test)
acc2=accuracy_score(y_test,predict2)
acc2
```
